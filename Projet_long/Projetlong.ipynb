### Model CNN actualiser

```python
### Model CNN

from keras.wrappers.scikit_learn import KerasClassifier
from sklearn.model_selection import cross_val_score
from keras.models import Sequential
from keras.layers import Dense, BatchNormalization
from tensorflow.keras.optimizers import Adam
from keras import activations
```

```python
def CNN_model_2(nb_classes = 2, dropoutRate=0.5, input_shape = (256,32,1), lr = 0.001) : 
    opt = Adam(learning_rate=lr)


    model = Sequential()
    # Layer 1 = input 
    # Layer 2
    model.add(Conv2D(64, kernel_size=(5,5), strides=(1,1), padding='same', activation='relu', input_shape=input_shape))
    model.add(Dense(64,activation='relu'))
    model.add(Dropout(dropoutRate))

    # Layer 3
    model.add(Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu'))
    model.add(BatchNormalization())
    # Apply activation.
    model.add(Dense(64,activation='relu'))

    # Layer 4
    model.add(MaxPooling2D((2,2), padding='same'))
    
    # Layer 5
    model.add(Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu'))
    model.add(Dense(64,activation='relu'))
    model.add(Dropout(dropoutRate))
    
    # Layer 6
    model.add(MaxPooling2D((2,2), padding='same'))
    
    # Layer 7
    model.add(Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu'))
    model.add(Dense(64,activation='relu'))
    model.add(Dropout(dropoutRate))
    
    # Layer 8
    model.add(MaxPooling2D((2,2), padding='same'))
    
    # Layer 9
    model.add(Conv2D(128, kernel_size=(3,3), strides=(1,1), padding='same', activation='relu'))
    model.add(Dense(64,activation='relu'))
    model.add(Dropout(dropoutRate))
    
    # Layer 10
    model.add(MaxPooling2D((2,2), padding='same'))
    
    # Layer 11
    model.add(Flatten())
    model.add(Dense(nb_classes, activation='softmax'))
    
    # Output
    return model



```

```python
Model_2 = CNN_model_2(nb_classes = 2, dropoutRate=0.5, input_shape = (256,32,1), lr=0.0001)
```

```python
Model_2.compile(loss='binary_crossentropy', optimizer=opt, metrics=['accuracy'])
```

#### Avec EEGNET

```python
def EEGNet(nb_classes = 2, Chans = 256, Samples = 32,K =16 , F1 = 16, 
             D = 2, F2 = 32):
    
    input1   = Input(shape = (Chans, Samples, 1))
    
    #############################################
    #############################################
    Sec1       = Conv2D(F1, (1, K), padding = 'same',input_shape = (Chans, Samples, 1),use_bias = False)(input1)
    Sec1       = BatchNormalization()(Sec1)
    Sec1       = Dropout(0.5)(Sec1)
    
    Sec2       = DepthwiseConv2D((32, 1), padding = 'valid',depth_multiplier = D,activation='linear')(Sec1)
    Sec2       = BatchNormalization()(Sec2)
    Sec2       = Activation('elu')(Sec2)
    Sec2       = AveragePooling2D((1, 4))(Sec2)
    Sec2       = Dropout(0.5)(Sec2)
    
    Sec3       = SeparableConv2D(F2, (1,16),padding = 'same',activation='linear')(Sec2)
    Sec3       = BatchNormalization()(Sec3)
    Sec3       = Activation('elu')(Sec3)
    Sec3       = AveragePooling2D((1, 8))(Sec3)
    Sec3       = Dropout(0.25)(Sec3)
        
    flatten      = Flatten(name = 'flatten')(Sec3)
    
    dense        = Dense(nb_classes, name = 'dense',activation='softmax')(flatten)
    #softmax      = Activation('softmax', name = 'softmax')(dense)
    
    return Model(inputs=input1, outputs=dense)
```

```python
import tensorflow
model_EEG = EEGNet()
model_EEG.compile(loss=tensorflow.keras.losses.BinaryCrossentropy(),optimizer=Adam(lr=0.0001), metrics=["accuracy"])
```
